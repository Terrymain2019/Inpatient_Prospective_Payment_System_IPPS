# Renaming columns on the table to eliminate white space

```python
drgs_columns = drgs.rename(columns={"DRG Definition": "DRG", "Provider Id": "Provider_Id", "Provider Name": "Provider_Name", "Provider Street Address": "Address", "Provider City": "City", "Provider State": "State", "Provider Zip Code": "Zip_Code", "Hospital Referral Region Description": "Region", " Total Discharges ": "Discharges", " Average Covered Charges ": "Avg_Coverage_Charges", " Average Total Payments ": "Avg_Total_Payments", "Average Medicare Payments": "Avg_Medicare_Payments"})
drgs_columns.head()
```

# Analyzing DRGs categories by filtering the highest: #of Discharges, Avg Coverage Charges, Avg Total Payments & Avg Medicare Payments.

```python
classes_drgs = drgs_columns.groupby('DRG').sum()
classes_drgs
```

```python
discharges_drgs = classes_drgs.sort_values('Discharges', ascending=False)
discharges_drgs.head()
```

```python
discharges_drgs = classes_drgs.sort_values('Discharges', ascending=True)
discharges_drgs.head()
```

```python
Coverage_Charges = classes_drgs.sort_values('Avg_Coverage_Charges', ascending=False)
Coverage_Charges
```

```python
Total_Payments = classes_drgs.sort_values('Avg_Total_Payments', ascending=False)
Total_Payments
```

```python
Medicare_Payments = classes_drgs.sort_values('Avg_Medicare_Payments', ascending=False)
Medicare_Payments
```
